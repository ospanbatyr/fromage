```python
import os
import re

def preprocess_text_reports(file_paths):
    processed_reports = []
    for file_path in file_paths:
        with open(file_path, 'r') as f:
            text = f.read()

        print(text)
        
        # Remove unnecessary and insensible parts
        text = re.sub(r"EXAMINATION:.*", "", text)  # Remove EXAMINATION line
        text = re.sub(r"WET READ:.*", "", text)  # Remove WET READ line
        text = re.sub(r"FINAL REPORT", "", text)  # Remove FINAL REPORT line
        text = re.sub(r"STUDY:.*", "", text)  # Remove STUDY line
        text = re.sub(r"COMPARISON:.*", "", text)  # Remove COMPARISON section
        text = re.sub(r"TECHNIQUE:.*", "", text)  # Remove TECHNIQUE section
        text = re.sub(r"_+", "", text)  # Remove underscores

        # Clean up excessive newlines and spaces
        text = re.sub(r"\s\s+", " ", text)
        text = re.sub(r" +", " ", text)
        text = text.strip()

        processed_reports.append(text)

    return processed_reports
```

```python
PATH = "/datasets/mimic/physionet.org/files/mimic-cxr/2.0.0/files/p10/p10251182"
paths = os.listdir("/datasets/mimic/physionet.org/files/mimic-cxr/2.0.0/files/p10/p10251182")
file_paths = []
for file_path in paths:
    if file_path.endswith(".txt"):
        file_paths.append(os.path.join(PATH,file_path))
```
